1. Stationarity Check
   We'll use the Augmented Dickey-Fuller (ADF) test and plot rolling statistics to test for stationarity. We'll perform this check for all three datasets.

   Sales Data Stationarity Check
   Let's begin by plotting the rolling mean and variance for the sales data and then perform the ADF test.

Interpretation:
Sales Quantity:

The ADF test indicates that the series is stationary.
The KPSS test suggests that the series is likely stationary.
Google Clicks:

The ADF test indicates that the series is non-stationary.
The KPSS test suggests that the series is non-stationary.
Facebook Impressions:

The ADF test indicates that the series is stationary.
The KPSS test suggests that the series is likely stationary.

1.(i). Sales Quantity

      ADF Test Statistic: -4.45
      ADF p-value: 0.00025 (indicating strong evidence against the null hypothesis of non-stationarity)
      KPSS Test Statistic: 0.323
      KPSS p-value: 0.1 (indicating weak evidence against the null hypothesis of stationarity)

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, kpss

# Load and process data
sales_data = pd.read_excel('C:\\Users\\sandi\\Downloads\\Hello\\Day 2\\ProductA.xlsx')
sales_data['Day Index'] = pd.to_datetime(sales_data['Day Index'])
sales_data.set_index('Day Index', inplace=True)

# Plot the time series
plt.figure(figsize=(10, 6))
plt.plot(sales_data.index, sales_data['Quantity'], label='Sales Quantity')
plt.title('Sales Quantity Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.show()

# Perform ADF and KPSS tests
sales_adf = adfuller(sales_data['Quantity'])
sales_kpss = kpss(sales_data['Quantity'], regression='c')

# Print results
print("Sales Quantity ADF Test:")
print(f"ADF Test Statistic: {sales_adf[0]}")
print(f"p-value: {sales_adf[1]}")
print(f"Critical Values: {sales_adf[4]}\n")

print("Sales Quantity KPSS Test:")
print(f"KPSS Test Statistic: {sales_kpss[0]}")
print(f"p-value: {sales_kpss[1]}")
print(f"Critical Values: {sales_kpss[3]}")
```

3.	Seasonality: The sales data exhibits seasonality (e.g., weekly, monthly, or yearly patterns). This is common in retail 
    data due to holidays, promotions, and other recurring events.
    Your findings with charts, plots and evidences to help in proving the hypothesis right or wrong

3(i). Sales Quantity Over Time
        Observation: The sales data over time was plotted, showing daily sales quantities.
        Trend Analysis: Seasonal decomposition was performed, revealing underlying trend, seasonal, and residual components.
        Findings: The seasonal component indicates if there are recurring patterns in sales, such as weekly or monthly fluctuations. The presence of significant patterns would confirm seasonality in sales.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the data
product_sales = pd.read_excel('C:\\Users\\sandi\\Downloads\\Hello\\Day 2\\ProductA.xlsx')

# Convert the 'Day Index' column to datetime
product_sales['Day Index'] = pd.to_datetime(product_sales['Day Index'])

# Set the 'Day Index' as the index
product_sales.set_index('Day Index', inplace=True)

# Perform seasonal decomposition
decomposition_sales = seasonal_decompose(product_sales['Quantity'], model='additive', period=30)

# Plot the decomposed components
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(14, 12))

ax1.plot(decomposition_sales.observed, label='Observed')
ax1.set_ylabel('Observed')
ax1.legend()

ax2.plot(decomposition_sales.trend, label='Trend', color='green')
ax2.set_ylabel('Trend')
ax2.legend()

ax3.plot(decomposition_sales.seasonal, label='Seasonal', color='orange')
ax3.set_ylabel('Seasonal')
ax3.legend()

ax4.plot(decomposition_sales.resid, label='Residual', color='red')
ax4.set_ylabel('Residual')
ax4.legend()

plt.tight_layout()
plt.show()

```

5.	Google Clicks Effect: An increase in Google clicks is associated with an increase in sales. This tests the effectiveness of online advertising.
Your findings with charts, plots and evidences to help in proving the hypothesis right or wrong

Loading Data: The script loads sales data and Google clicks data from Excel files.
Data Preparation: It converts the 'Day Index' columns to datetime format and merges the two datasets on 'Day Index'.
Time Series Plot: The script plots sales quantity and Google clicks over time to visualize their trends.
Correlation Analysis: It calculates and prints the correlation coefficient between Google clicks and sales.
Scatter Plot with Regression Line: Finally, the script creates a scatter plot with a regression line to visualize the relationship between Google clicks and sales.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the sales data
product_sales = pd.read_excel('C:\\Users\\sandi\\Downloads\\Hello\\Day 2\\ProductA.xlsx')
# Load the Google clicks data
google_clicks = pd.read_excel('C:\\Users\\sandi\\Downloads\\Hello\\Day 2\\ProductA_google_clicks.xlsx')

# Convert 'Day Index' to datetime
product_sales['Day Index'] = pd.to_datetime(product_sales['Day Index'])
google_clicks['Day Index'] = pd.to_datetime(google_clicks['Day Index'])

# Merge the datasets on 'Day Index'
merged_data = pd.merge(product_sales, google_clicks, on='Day Index')

# Plot the sales and Google clicks data
plt.figure(figsize=(14, 7))
plt.plot(merged_data['Day Index'], merged_data['Quantity'], label='Sales Quantity', color='blue')
plt.plot(merged_data['Day Index'], merged_data['Clicks'], label='Google Clicks', color='green')
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Sales Quantity and Google Clicks Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the correlation coefficient
correlation = merged_data['Quantity'].corr(merged_data['Clicks'])
print(f"Correlation coefficient between Google clicks and sales: {correlation:.2f}")

# Create a scatter plot with a regression line
plt.figure(figsize=(14, 7))
sns.regplot(x='Clicks', y='Quantity', data=merged_data, scatter_kws={'color': 'green'}, line_kws={'color': 'red'})
plt.xlabel('Google Clicks')
plt.ylabel('Sales Quantity')
plt.title('Scatter Plot of Google Clicks vs. Sales Quantity')
plt.grid(True)
plt.show()
```

7.	Lagged Effects: Past values of sales, clicks, or impressions have a significant impact on current sales. This could indicate a delayed response to marketing efforts or a carryover effect from previous periods.
Your findings with charts, plots and evidences to help in proving the hypothesis right or wrong.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the provided Excel files
sales_data = pd.read_excel('C:\\Users\\sandi\\Downloads\\Hello\\Day 2\\ProductA.xlsx')
google_clicks_data = pd.read_excel('C:\\Users\\sandi\\Downloads\\Hello\\Day 2\\ProductA_google_clicks.xlsx')
fb_impressions_data = pd.read_excel('C:\\Users\\sandi\\Downloads\\Hello\\Day 2\\ProductA_fb_impressions.xlsx')

# Renaming columns for clarity
sales_data.columns = ['Date', 'Sales']
google_clicks_data.columns = ['Date', 'Clicks']
fb_impressions_data.columns = ['Date', 'Impressions']

# Merging the datasets on 'Date'
merged_data = sales_data.merge(google_clicks_data, on='Date').merge(fb_impressions_data, on='Date')

# Plotting Sales, Clicks, and Impressions over time
fig, axs = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

axs[0].plot(merged_data['Date'], merged_data['Sales'], label='Sales', color='blue')
axs[0].set_ylabel('Sales')
axs[0].legend()

axs[1].plot(merged_data['Date'], merged_data['Clicks'], label='Google Clicks', color='green')
axs[1].set_ylabel('Clicks')
axs[1].legend()

axs[2].plot(merged_data['Date'], merged_data['Impressions'], label='Facebook Impressions', color='red')
axs[2].set_ylabel('Impressions')
axs[2].set_xlabel('Date')
axs[2].legend()

plt.suptitle('Sales, Google Clicks, and Facebook Impressions Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Creating lagged variables for clicks and impressions
for lag in range(1, 8):  # lag from 1 to 7 days
    merged_data[f'Clicks_Lag_{lag}'] = merged_data['Clicks'].shift(lag)
    merged_data[f'Impressions_Lag_{lag}'] = merged_data['Impressions'].shift(lag)

# Dropping rows with NaN values resulting from the lagging process
lagged_data = merged_data.dropna()

# Calculating correlations
correlation_data = {
    'Lag': [],
    'Clicks_Correlation': [],
    'Impressions_Correlation': []
}

for lag in range(1, 8):
    correlation_data['Lag'].append(lag)
    correlation_data['Clicks_Correlation'].append(lagged_data['Sales'].corr(lagged_data[f'Clicks_Lag_{lag}']))
    correlation_data['Impressions_Correlation'].append(lagged_data['Sales'].corr(lagged_data[f'Impressions_Lag_{lag}']))

correlation_df = pd.DataFrame(correlation_data)

# Plotting the correlations
plt.figure(figsize=(10, 6))
plt.plot(correlation_df['Lag'], correlation_df['Clicks_Correlation'], marker='o', label='Clicks Correlation')
plt.plot(correlation_df['Lag'], correlation_df['Impressions_Correlation'], marker='o', label='Impressions Correlation')
plt.xlabel('Lag (days)')
plt.ylabel('Correlation with Sales')
plt.title('Correlation of Sales with Lagged Clicks and Impressions')
plt.legend()
plt.grid(True)
plt.show()

print(correlation_df)
```
